## EDA (Exploratory Data Analysis)

```python
import plotly.express as px
```

### User Data

```python
# Get User Data
print(f"Number of users: {len(df['user_id'].unique())}")
user_df = df[['user_id', 'age', 'sex', 'country']].copy()
user_df.drop_duplicates(inplace=True)
user_df
```

#### User Counts by Sex

```python
# Plot User Counts by Sex
fig = px.histogram(user_df, x='sex', title='User Counts by Sex')
fig.show()
```

- The vast majority of users (82%) are female

#### Top 10 Countries by User Counts

```python
# Get Top 10 countries df
country_df = user_df.groupby('country').count()['user_id'].reset_index()
country_df['count'] = country_df['user_id']
country_df = country_df.sort_values('count', ascending=False)[:10]
# Plot Top 10 Countries by User Counts
fig = px.pie(country_df, values='count', names='country', 
             title='Top 10 Countries by User Counts',
             )
fig.show()
```

- Top countries are: US, followed by Great Britain, Canada, Australia, then New Zealand and others

#### User Counts by Age

```python
df['age'].value_counts()
```

```python
# Get invalid ages data
user_df.loc[(df['age'] > 100) | (df['age'] <= 0)]
```

- There are invalid ages (> 100 or <= 0), so let's remove them.

```python
# Keep only valid ages
valid_age_user_df = user_df.loc[(user_df['age'] <= 100) 
                                & (user_df['age'] > 0)].copy()
print(f"Number of users: {len(valid_age_user_df['user_id'].unique())}")
# Plot User Counts by Age
fig = px.histogram(valid_age_user_df, x='age', nbins=20, 
                   title='User Counts by Age')
fig.show()
```

- Most users (7,475 or 18%) are 25-29
- There are a good amount of users (380) aged 0-4
    - Perhaps these users are parents tracking their baby's health

### Check-In Counts by Month

```python
import datetime
```

```python
# Get user check-in dates df
checkin_df = df[['user_id', 'checkin_date']].drop_duplicates()
checkin_df['checkin_date'] = pd.to_datetime(checkin_df['checkin_date'])
checkin_df = checkin_df.set_index('checkin_date')
checkin_df
```

```python
# Get monthly check-in counts
monthly_checkin_df = checkin_df.resample('ME').count()
monthly_checkin_df['count'] = monthly_checkin_df['user_id']
monthly_checkin_df = monthly_checkin_df.reset_index()
monthly_checkin_df
# Plot Check-In Counts by Month
fig = px.line(monthly_checkin_df, x='checkin_date', y='count', title='Check-In Counts by Month')
fig.show()
```

- The highest number of check-ins occurred in Jan 2019
    - This may be due to excitement for the new year and New Year's resolutions
- There is a dropoff in check-in counts in Dec 2019
    - This may be due to the dataset ending with only partial data for the final month

### Trackables

#### Trackable Types

```python
# Get Trackable Types df
trackable_types_df = df.groupby('trackable_type').count()['user_id'].reset_index()
trackable_types_df['count'] = trackable_types_df['user_id']
trackable_types_df = trackable_types_df.sort_values('count', ascending=False)
# Plot Trackable Types Breakdown
fig = px.pie(trackable_types_df, values='count', names='trackable_type', 
             title='Trackable Types Breakdown',
             )
fig.show()
```

- Top trackable types are: Symptom, followed by Weather, Condition, Treatment, Food,  then Tag and HBI

#### Trackable Names

```python
# Get Top 10 Trackable names df
trackable_names_df = df.groupby(['trackable_type', 'trackable_name']).count()['user_id'].reset_index()
trackable_names_df['count'] = trackable_names_df['user_id']
trackable_names_df = trackable_names_df.sort_values('count', ascending=False)
trackable_names_df
```

- Let's drop Weather trackables, which are pulled via API and not user input

```python
trackable_names_df = trackable_names_df.drop(trackable_names_df.loc[trackable_names_df['trackable_type']=='Weather'].index)
trackable_names_df = trackable_names_df.sort_values('count', ascending=False)[:15]
trackable_names_df
```

```python
# Plot Top 15 Trackable Names Breakdown
fig = px.pie(trackable_names_df, values='count', names='trackable_name', 
             title='Top 15 Trackable Names Breakdown',
             )
fig.show()
```

- The top trackable names are: Headache, followed by Anxiety, Fatigue, Depression, and then Nausea and others
